# This notebook simulates throwing a regular dice, with 6 faces, 10, 1000, 10000 and 100000 times.
### This is a simple exercise to analyse the probability of getting each result, from 1 to 6.

- It was written by Aphonso Rafael

```python
import random
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def throw_dice_prob_chart(times):
    try:
        del r
    except:
        pass
    list = []
    n = times

    for i in range(n):
        r = random.randint(1, 6)
        list.append(r)

    result_df = pd.DataFrame(list)
    result_df = result_df.reset_index()
    result_df.columns = ['Throw', 'Result']
    result_df['Throw'] = result_df['Throw'] + 1

    result_df['Prob_1'] = (result_df['Result'] == 1).cumsum() / result_df['Throw']
    result_df['Prob_2'] = (result_df['Result'] == 2).cumsum() / result_df['Throw']
    result_df['Prob_3'] = (result_df['Result'] == 3).cumsum() / result_df['Throw']
    result_df['Prob_4'] = (result_df['Result'] == 4).cumsum() / result_df['Throw']
    result_df['Prob_5'] = (result_df['Result'] == 5).cumsum() / result_df['Throw']
    result_df['Prob_6'] = (result_df['Result'] == 6).cumsum() / result_df['Throw']

    plt.figure(num=3, figsize=(15, 10))
    plt.plot(result_df['Throw'], result_df['Prob_1'], color='blue', linewidth=1.0, label='1')
    plt.plot(result_df['Throw'], result_df['Prob_2'], color='red', linewidth=1.0, label='2')
    plt.plot(result_df['Throw'], result_df['Prob_3'], color='yellow', linewidth=1.0, label='3')
    plt.plot(result_df['Throw'], result_df['Prob_4'], color='black', linewidth=1.0, label='4')
    plt.plot(result_df['Throw'], result_df['Prob_5'], color='green', linewidth=1.0, label='5')
    plt.plot(result_df['Throw'], result_df['Prob_6'], color='orange', linewidth=1.0, label='6')

    plt.axhline(linestyle='--',
                y=1/6,
                xmin=0,
                color='gray')

    plt.xlabel("Throws")
    plt.ylabel("Probability")
    plt.legend(loc='upper right')

    plt.show()
```

### Throwing 10 times

```python
throw_dice_prob_chart(10)
```

### Throwing 1000 times

```python
throw_dice_prob_chart(1000)
```

### Throwing 10000 times

```python
throw_dice_prob_chart(10000)
```

### Throwing 100000 times

```python
throw_dice_prob_chart(100000)
```
